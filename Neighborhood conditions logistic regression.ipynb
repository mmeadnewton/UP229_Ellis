```python
data=parcelsjoindf4[['units_total','yearbuilt_avg',
                    'median_income_2014','median_rent_2014','part_black_2014',
                    'part_latx_2014', 'part_female_head_2014', 'part_hh_kids_2014',
                    'part_rented_2014', 'part_unemployed_2014', 'part_vacant_2014',
                    'part_college_2014','dist_to_nearest_highrenttract',
                     'ellis_status']]

Y =data['ellis_status']
X =data.drop(['ellis_status'],1)
X = sm.add_constant(X)
log_reg = sm.Logit(Y, X).fit()

print(log_reg.summary())
```
