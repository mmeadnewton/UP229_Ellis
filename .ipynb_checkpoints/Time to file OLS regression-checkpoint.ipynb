```python
data=ellissamplefinaldf[['units_total','sqft_total','yearbuilt_avg','n_ll_properties', 'n_ll_units', 'landlord_iscorp',
                    'median_income_2019','median_rent_2019','median_value_2019','part_black_2019',
                    'part_latx_2019', 'part_white_2019', 'part_female_head_2019', 'part_hh_kids_2019',
                    'part_rented_2019', 'part_unemployed_2019', 'part_poverty_2019', 'part_vacant_2019',
                    'part_college_2019','dist_to_nearest_highrenttract', 'dist_to_nearest_highchangetract','pctbuilt','pctsold','time_to_file']]


Y =data['time_to_file']
X =data.drop(['time_to_file'],1)
X = sm.add_constant(X)
model = sm.OLS(Y,X)
results = model.fit()
results.params
print(results.summary())
```
